# Import libraries and load dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
df=pd.read_csv(r"C:\Users\cheta\OneDrive\Documents\1 Python (AI_ML)\1project dataset\7 Student_Performance.csv")
```

```python
df.head()
```

# Split data into train and test

```python
X=df.drop("Performance Index",axis=1)
y=df['Performance Index']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=11)
```

# Encoding dataset

```python
ohe=OrdinalEncoder()
```

```python
X_train["Extracurricular Activities"]=ohe.fit_transform(X_train[["Extracurricular Activities"]])
```

```python
X_test["Extracurricular Activities"]=ohe.transform(X_test[["Extracurricular Activities"]])
```

# Train the model (Decision Tree )

```python
decision_tree=DecisionTreeRegressor()
```

```python
decision_tree.fit(X_train,y_train)
```

```python
y_pred=decision_tree.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
decision_tr=DecisionTreeRegressor(max_depth=4)
```

```python
decision_tr.fit(X_train,y_train)
```

```python
y_pred=decision_tr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
# mean_squared_error(y_test,y_pred)
mean_absolute_error(y_test,y_pred)
```
